```python
def plot(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(7,7), autopct='%1.2f%%', fontsize = 10)
    ax.set_title(variable + ' \n', fontsize = 10)
    return n.round(dataframe_pie/df.shape[0]*100,2)
```

```python
del df['Sl.No']
```

```python
x1 = df.drop(labels='Grade', axis=1)
y1 = df.loc[:,'Grade']
```

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

ros =RandomOverSampler()
x,y=ros.fit_resample(x1,y1)
print("OUR DATASET COUNT         : ", Counter(y1))
print("OVER SAMPLING DATA COUNT  : ", Counter(y))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42, stratify=y)
```

```python
print("NUMBER OF TRAIN DATASET    : ", len(x_train))
print("NUMBER OF TEST DATASET      : ", len(x_test))
print("TOTAL NUMBER OF DATASET    : ", len(x_train)+len(x_test))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, hamming_loss
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC = RandomForestClassifier()
RFC.fit(x_train,y_train)
predicted = RFC.predict(x_test)
```

```python
cr = classification_report(y_test,predicted)
print('\n THE CLASSIFICATION REPORT OF RANDOM FOREST CLASSIFIER:\n\n',cr)
```

```python
cm = confusion_matrix(y_test,predicted)
print('\nTHE CONFUSION MATRIX SCORE OF RANDOM FOREST CLASSIFIER:\n\n',cm)
print("\n")
```

```python
accuracy = cross_val_score(RFC, x, y, scoring='accuracy')
print('\nTHE CROSS VALIDATION TEST RESULT OF ACCURACY :\n\n', accuracy*100)
```

```python
a = accuracy_score(y_test,predicted)
print("\nTHE ACCURACY SCORE OF RANDOM FOREST CLASSIFIER IS :",a*100)
```

```python
hl = hamming_loss(y_test,predicted)
print("\nTHE HAMMING LOSS OF RANDOM FOREST CLASSIFIER IS :",hl*100)
```

```python
def plot_confusion_matrix(cm, title='THE CONFUSION MATRIX SCORE OF RANDOM FOREST CLASSIFIER\n', cmap=plt.cm.viridis):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()

cm1=confusion_matrix(y_test, predicted)
print('\nTHE CONFUSION MATRIX SCORE OF RANDOM FOREST CLASSIFIER:\n')
print(cm,'\n')
plot_confusion_matrix(cm)
```

```python
import matplotlib.pyplot as plt
df2 = p.DataFrame()
df2["y_test"] = y_test
df2["predicted"] = predicted
df2.reset_index(inplace=True)
plt.figure(figsize=(20, 5))
plt.plot(df2["predicted"][:100], marker='x', linestyle='dashed', color='red')
plt.plot(df2["y_test"][:100],  marker='o', linestyle='dashed', color='green')
plt.show()
```
